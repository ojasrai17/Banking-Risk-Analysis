```python
bins=[0,100000,300000,float('inf')]
labels=['low','medium','hard']
df['Income Band']=pd.cut(df['Estimated Income'],bins=bins,labels=labels,right=False)

```

```python
# Explain the distribution of unique categories in categorical columns
categorical_cols=df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']]
for i in categorical_cols:
    print(f"Value counts for {i}")
    display(df[i].value_counts())
```

# Univariate Analysis

```python
for i,pred in enumerate(categorical_cols):
    plt.figure(i)
    sns.countplot(data=df,x=pred)
```

# Bivariate Analysis

```python
for i,pred in enumerate(categorical_cols):
    plt.figure(i)
    sns.countplot(data=df,x=pred,hue='Nationality')
```

```python
for col in categorical_cols:
    if col=='Occupation':
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of count')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
    
        
```

# Numerical Analysis


```python
numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending' ]
# univariate analysis and visualizations
for i,col in enumerate(numerical_cols):
    plt.figure(figsize=(15,10))
    plt.subplot(3,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
    plt.tight_layout()
    plt.show()
    
```

# Heatmap

```python
numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending' ]
correlation_matrix=df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
